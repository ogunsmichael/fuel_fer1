## Answer to question 6

```python
# Shows the number of null value
null_data = Fuel_Data.isnull().sum()
null_data
```

```python
# Counts the total number of enries in the Column
null_data_sum = Fuel_Data.groupby('fuel_unit')['fuel_unit'].count().sum()
null_data_sum
```

```python
# The formula for  percentage of missing value
null_percent = null_data * 100 / null_data_sum
null_percent * 100
```
